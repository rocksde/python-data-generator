```python
from datetime import datetime
import requests
import time
import random

##class for data_generation
def data_generation():
    satellite = random.choice( ['FM09', 'FM05', 'FM07', 'FM10', 'FM12'] )
    txt1 = random.choice( ['Aa', 'Bb', 'Cc', 'Dd', 'Ee'] )
    txt2 = random.choice( ['Rtn', 'Fwd'] )
    txt3 = random.choice( ['Ka', 'Ku', 'K'] )
    num1 = random.randint(500,1500)
    num2 = random.randint(0,350)
    num3 = random.randint(0,10)
    ts_unix = int(time.time())
    timestamp = datetime.now().isoformat()

    return [satellite, txt1, txt2, txt3, num1, num2, num3, ts_unix, timestamp]

if __name__ == '__main__':
    # change URL to the location that the payload should be send to 
    REST_API_URL = 'http://localhost:8008/post_satdata'
  
    while True:
        for i in range(1):
            data_record = data_generation()

        # set the header record
        vHeader = ["satellite", "txt1", "txt2", "txt3", "num1", "num2", "num3", "ts_unix", "timestamp"]
        data_df = dict(zip(vHeader, data_record))
        
        # Post the data to URL
        req = requests.post(REST_API_URL, json=data_df)
        print(data_df)

        #wait time until next record
        time.sleep(0)
```
